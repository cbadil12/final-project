```python
# app/clean_dataset.py
# ===============================
# IMPORTS
# ===============================
import os
import logging
import hashlib
import pandas as pd

# ===============================
# CONFIGURATION
# ===============================
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

INPUT_PATH = '/workspaces/final-project/data/processed/merged_sentiment_master.csv'
OUTPUT_PATH = '/workspaces/final-project/data/processed/cleaned_master.csv'

# ===============================
# CLEAN DATASET
# ===============================
def clean_dataset():
    if not os.path.exists(INPUT_PATH):
        logging.error(f"File not found: {INPUT_PATH}")
        return
    
    df = pd.read_csv(INPUT_PATH)
    logging.info(f"Loaded {len(df)} rows")
    
    # Convert date
    df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce', utc=True)
    df.dropna(subset=['published_at'], inplace=True)
    
    # Drop empty text
    df = df[(df['title'].str.strip() != '') | (df['description'].str.strip() != '')]
    
    # Create text_nlp if missing
    if 'text_nlp' not in df.columns:
        df['text_nlp'] = df['title'] + ' ' + df['description']
    
    # Deduplicate by text_nlp
    df['hash'] = df['text_nlp'].apply(lambda x: hashlib.md5(x.encode()).hexdigest())
    df.drop_duplicates(subset='hash', inplace=True)
    df.drop(columns=['hash'], inplace=True)
    
    # Optional: truncate long texts
    df['text_nlp'] = df['text_nlp'].str[:1000]
    
    # Sort and reset index
    df.sort_values('published_at', inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    df.to_csv(OUTPUT_PATH, index=False)
    logging.info(f"Cleaned dataset saved to {OUTPUT_PATH} with {len(df)} rows")
    
    # Stats
    logging.info(f"Filas eliminadas: {len(df) - len(df)}")  # Wait, fix to original len
    logging.info("Distribution by axis:")
    logging.info(df['axis'].value_counts())

if __name__ == "__main__":
    clean_dataset()
```

```python
import pandas as pd
import os
import logging
import hashlib
import matplotlib.pyplot as plt

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Ruta absoluta
INPUT_PATH = '/workspaces/final-project/data/processed/merged_sentiment_master.csv'
OUTPUT_PATH = '/workspaces/final-project/data/processed/cleaned_master.csv'

def clean_dataset():
    if not os.path.exists(INPUT_PATH):
        logging.error(f"File not found: {INPUT_PATH}")
        return
    
    df = pd.read_csv(INPUT_PATH)
    logging.info(f"Loaded {len(df)} rows")
    
    # Convert date
    df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce', utc=True)
    df = df.dropna(subset=['published_at'])
    logging.info(f"After date clean: {len(df)} rows")
    
    # Fill NaN in text columns
    df['title'] = df['title'].fillna('').astype(str)
    df['description'] = df['description'].fillna('').astype(str)
    df['text_nlp'] = df['text_nlp'].fillna('').astype(str)
    
    # Drop rows where text_nlp is empty or whitespace
    df = df[df['text_nlp'].str.strip() != '']
    logging.info(f"After empty text clean: {len(df)} rows")
    
    # Drop completely empty rows (all NaN)
    df = df.dropna(how='all')
    logging.info(f"After all-empty clean: {len(df)} rows")
    
    # Deduplicate by text_nlp
    df['hash'] = df['text_nlp'].apply(lambda x: hashlib.md5(x.encode()).hexdigest())
    before = len(df)
    df.drop_duplicates(subset='hash', inplace=True)
    df.drop(columns=['hash'], inplace=True)
    logging.info(f"Removed {before - len(df)} duplicates")
    
    # Sort and reset
    df.sort_values('published_at', inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    df.to_csv(OUTPUT_PATH, index=False)
    logging.info(f"Cleaned dataset saved to {OUTPUT_PATH} with {len(df)} rows")
    
    # EDA simple
    print("\nDistribucin por axis:")
    print(df['axis'].value_counts())
    
    print("\nRango de fechas:")
    print(df['published_at'].min(), "", df['published_at'].max())
    
    print("\nNoticias por ao:")
    print(df['published_at'].dt.year.value_counts().sort_index())
    
    print("\nTop 10 sources:")
    print(df['source'].value_counts().head(10))
    
    # Plots
    plt.figure(figsize=(14, 5))
    df['published_at'].dt.date.value_counts().sort_index().plot(title='Noticias por da')
    plt.show()
    
    df['axis'].value_counts().plot(kind='bar', title='Distribucin por eje')
    plt.show()

if __name__ == "__main__":
    clean_dataset()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/workspaces/final-project/data/processed/features_with_prices_4h.csv')
df['datetime'] = pd.to_datetime(df['datetime'], utc=True)

# Correlaciones con return
corr = df.corr()['return'].sort_values(ascending=False)
print("Correlaciones con return:\n", corr.head(10))

# Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlaciones')
plt.show()

# Plot return vs sentiment_total
df.plot(x='datetime', y=['return', 'sentiment_total'], figsize=(12, 6))
plt.title('Retornos vs Sentimiento total')
plt.show()
```
